# Fast Text

```python
import os
import pandas as pd
from tqdm.auto import tqdm
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from gensim.models import FastText

from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

Prepare Corpus

```python
df = pd.read_csv("/content/drive/My Drive/Tesis/re_dataset.csv", encoding='Windows-1252')
df.head()
```

```python
sentences = [word_tokenize(Tweet.lower()) for Tweet in tqdm(df.Tweet)]
sentences[:5]
```

Training

```python
model = FastText(sentences, size=100, window=3, min_count=3, sg=0, workers=4, iter=1000)
```

```python
os.makedirs("model/", exist_ok=True)
model.save("model/model.fasttext")
```

Load Model

```python
model = FastText.load("model/model.fasttext")
```

Continue training

```python
"Kaum cebong kapir udah keliatan dongoknya dari awal tambah dongok lagi hahahah".lower().split()
```

```python
contoh_data = [
    ['deklarasi',
      'pilkada',
      '2018',
      'aman',
      'dan',
      'anti',
      'hoax',
      'warga',
      'dukuh',
      'sari',
      'jabon'],
    ['kaum',
      'cebong',
      'kapir',
      'udah',
      'keliatan',
      'dongoknya',
      'dari',
      'awal',
      'tambah',
      'dongok',
      'lagi',
      'hahahah']
]
```

```python
model.train(contoh_data, total_examples=len(contoh_data), epochs=1)
```

```python
model.save("model/model.fasttext")
```

higher order visualization

```python
from umap import UMAP
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
X = UMAP().fit_transform(fasttext.vectors)
```

```python
df = pd.DataFrame(X, columns=["umap1", "umap2"])
df["text"] = fasttext.index2word
```

```python
df
```

```python
fig = px.scatter(df, x="umap1", y="umap2", text="text")
fig.update_traces(textposition='top center')
fig.update_layout(
    height=800,
    title_text='Reduced word2vec visualization'
)
fig.show()
```
